```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data_tgp_1.csv', sep=';')

data.head()
```

```python
data.describe()
```

```python
plt.figure(figsize=(10,8))

df=data[data['churn']==1]
plt.subplot(2,1,1)
plt.hist(df['vintage'],bins=20)
plt.xlabel('vintage (churn)')
plt.grid(linewidth=0.5)

df_2=data[data['churn']==0]
plt.subplot(2,1,2)
plt.hist(df_2['vintage'], bins=20)
plt.xlabel('vintage (non churn)')
plt.grid(linewidth=0.5)

plt.show()
```

```python
plt.figure(figsize=(15,8))

df=data[data['churn']==1]
plt.subplot(2,1,1)
sns.boxenplot(df['current_balance'])
plt.xlabel('balance (churn)')
plt.xlim(xmin=-10000,xmax=100000000)
labels,locations=plt.xticks()
plt.xticks(labels,(labels/100000).astype(int))

df_2=data[data['churn']==0]
plt.subplot(2,1,2)
sns.boxenplot(df_2['current_balance'])
plt.xlabel('balance (non churn)')
plt.xlim(xmin=-10000,xmax=100000000)
labels,locations=plt.xticks()
plt.xticks(labels,(labels/100000).astype(int))

plt.show()
```

```python
data.corr()
```

```python
gender_churn = pd.crosstab(data.gender,data.churn)
plt.figure()
gender_churn.plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(10,8))

df=data[data['churn']==1]
plt.subplot(2,1,1)
plt.hist(df['age'], bins=20)
plt.xlabel('age (churn)')
plt.grid(linewidth=0.5)

df_2=data[data['churn']==0]
plt.subplot(2,1,2)
plt.hist(df_2['age'], bins=20)
plt.xlabel('age (non churn)')
plt.grid(linewidth=0.5)

plt.show()
```

```python
nn = pd.crosstab(data.vintage,data.churn)
plt.figure()
nn.plot(kind='line')
plt.show()
```

```python
plt.figure()
sns.boxenplot(data=data, x='vintage', y='gender')
plt.show()
```

```python
cust_cat_churn = pd.crosstab(data.customer_nw_category,data.churn)
plt.figure()
cust_cat_churn.plot.bar()
plt.grid(linewidth=0.5)
plt.show()
```
