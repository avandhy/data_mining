#### Feature Reduction

A method for reducing the dimentionality of data (In our case study, we have 4 features)

If we have m-columns, we can reduce it to m or fewer columns, whilist retaining the essence of the original data

```python
# dimentionality reduction
from sklearn.decomposition import PCA
```

```python
# Create a PCA instance: pca
pca = PCA(n_components = 2)
pca_component = pca.fit_transform(X)

# Plot the explained variances
features = range(pca.n_components_)
pca_variance = pca.explained_variance_ratio_
pca_result = pd.DataFrame({'PCA Feature': features,'Variance (%)': pca_variance})
fig = px.bar(pca_result, x= pca_result.columns[0], y= pca_result.columns[1], text= pca_result.columns[1])
fig.update_layout(title_text= 'Variation per Principal Component')
fig.show()
```

PCA 1 merepresentasikan 92% dari informasi keseluruhan, sementara PCA 2 merepresentasikan 4% informasi keseluruhan

```python
print('Explained variation per principal component: {}'.format(pca_variance))
```

```python
# create dataframe consist of principal component
df_pca = pd.DataFrame(data = pca_component, columns = ['Principal Component-1', 'Principal Component-2'])
finalDf = pd.concat([df_pca, y], axis = 1)

# # Data Visualization after PCA
# sb.FacetGrid(df_pca, height= 6).map(plt.scatter, df_pca.columns[0], df_pca.columns[1])
# plt.show()
```

### K-Means

Ada 2 cara dalam mengimplementasikan K-Means di Python. Pertama menggunakan library sklearn. Kedua menggunakan scipy. Penggunaan k-means di library scipy akan dijelaskan di bagian Vector Quantization

Algoritma K-means menggunakan KMeans() pada library sklearn.cluster. Beberapa parameter yang bisa digunakan :


*   n_clusters : total klaster yang digunakan; default = 8

---

Sumber : https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html



```python
n_cluster = 11
kmean = KMeans(n_clusters= n_cluster)

# fitting model
kmean.fit(X_pca)

# mengambil label hasil klaster
result_kmeans = pd.DataFrame(X_pca.copy())
result_kmeans['Cluster'] = kmean.labels_
result_kmeans
```

```python
# ambil koordinat centroid
centers = kmean.cluster_centers_
print("Koordinat Centroid {}".format(centers))
```

```python
# plot clustering result
plt.figure(figsize=(12, 8))
sb.scatterplot(
    result_kmeans.iloc[:,0], result_kmeans.iloc[:,1],
    s = 120,
    hue= result_kmeans.Cluster, 
    # palette= ['g', 'r', 'b', 'y'])
    palette= sb.color_palette('hls', len(result_kmeans.Cluster.unique())))

# plot the centroids
centroids = kmean.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker = "*", s = 100, color = 'black')
# define plot title
plt.title('KMeans with ' + str(n_cluster) + ' Clusters')
plt.show()
```

### Cluster Evaluation

####Define function

```python
from scipy.stats import entropy as en
```

```python
def entropy_cluster(y_true, y_pred):
  # Membuat matriks kontingensi
  contingency_matrix = metrics.cluster.contingency_matrix(y_true, y_pred)
  # Menghitung nilai entropy
  entropy_list = []
  for i in range(len(contingency_matrix)):
    total = np.sum(contingency_matrix[:,i])
    a = []
    for j in range(len(contingency_matrix[:,i])):
      a.append(contingency_matrix[:,i][j]/total)
    entropy_list.append(round(en(a, base=2),4))
  # Mengembalikan nilai entropy
  return entropy_list
```

```python
def purity_cluster(y_true, y_pred):
  # Membuat matriks kontingensi
  contingency_matrix = metrics.cluster.contingency_matrix(y_true, y_pred)
  # Menghitung nilai purity
  purity_list = []
  for i in range(len(contingency_matrix)):
    total = np.sum(contingency_matrix[:,i])
    a = []
    for j in range(len(contingency_matrix[:,i])):
      a.append(contingency_matrix[:,i][j]/total)
    purity_list.append(round(max(a),4))
  # Mengembalikan nilai purity
  return purity_list
```

```python
def entropy_score(y_true, y_pred):
    # compute contingency matrix (also called confusion matrix)
    entropy_list = entropy_cluster(y_true, y_pred)
    contingency_matrix = metrics.cluster.contingency_matrix(y_true, y_pred)
    a= []
    total = np.sum(contingency_matrix)
    for i in range(len(contingency_matrix)):
      total_cluster = np.sum(contingency_matrix[:,i])
      p = total_cluster*entropy_list[i]/total
      a.append(p)
    print('Nilai entropy keseluruhan adalah {0:.4f} \nNilai entropy masing-masing klaster adalah {1}'.format(np.sum(a), entropy_list))
```

```python
def purity_score(y_true, y_pred):
    # compute contingency matrix (also called confusion matrix)
    purity_list = purity_cluster(y_true, y_pred)
    contingency_matrix = metrics.cluster.contingency_matrix(y_true, y_pred)
    a= []
    total = np.sum(contingency_matrix)
    for i in range(len(contingency_matrix)):
      total_cluster = np.sum(contingency_matrix[:,i])
      p = total_cluster*purity_list[i]/total
      a.append(p)
    print('Nilai purity keseluruhan adalah {0:.4f} \nNilai purity masing-masing klaster adalah {1}'.format(np.sum(a), purity_list))
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
n_minsupport = 0.1

# Menghasilkan association rules berdasarkan dataset
frequent_itemsets_fpgrowth = fpgrowth(codon_2, min_support= n_minsupport, use_colnames=True)
frequent_itemsets_fpgrowth.sort_values(by='support', ascending=False)
```

Want to filter by the support value?

```python
frequent_itemsets_fpgrowth[(frequent_itemsets_fpgrowth.support >= 0.05)].sort_values(by=['support'], ascending=False)
```

```python
# Let's say you are interested in rules derived from the frequent itemsets only if the level of confidence is above the 20 percent threshold
n_treshold = 0.2
rule_metrics = "confidence"

rules_result_fpgrowth = association_rules(frequent_itemsets_fpgrowth, metric= rule_metrics, min_threshold= n_treshold)

# urutkan berdasarkan nilai confidence
rules_result_fpgrowth.sort_values(by=[rule_metrics], ascending=False)
```

Not interested in confidence indicator? You can change it!

We can change the 'metric' parameter into our preferred indicator(e.g., confidence) and define the treshold by tune the 'min_treshold' parameter association_rules()

```python
# Let's say you are interested in rules derived from the frequent itemsets only if the level of confidence is above the 20 percent threshold
n_treshold = 1.2
rule_metrics = "lift"

rules_result_fpgrowth = association_rules(frequent_itemsets_fpgrowth, metric= rule_metrics, min_threshold= n_treshold)

# urutkan berdasarkan nilai confidence
rules_result_fpgrowth.sort_values(by=[rule_metrics], ascending=False)
```

```python
rules_result_fpgrowth = rules_result_fpgrowth[['antecedents', 'consequents', 'support', 'confidence']]
rules_result_fpgrowth
```

*   'support' is used to measure the frequency (often interpreted as significance or importance) of an itemset in a database
*   'confidence' is the probability of seeing the consequent in a transaction given that it also contains the antecedent
* 'lift' is commonly used to measure how much more often the antecedent and consequent of a rule A->C occur together than we would expect if they were statistically independent. If A and C are independent, the Lift score will be exactly 1
* 'leverage' computes the difference between the observed frequency of A and C appearing together and the frequency that would be expected if A and C were independent
* A high 'conviction' value means that the consequent is highly depending on the antecedent
